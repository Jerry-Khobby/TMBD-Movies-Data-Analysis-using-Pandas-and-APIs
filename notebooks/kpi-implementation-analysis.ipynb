# User-Defined Function

```python
# UDF to streamline ranking operations
def rank_movies(df: pd.DataFrame, metric: str, ascending: bool = False) -> pd.DataFrame:
    db = df.copy()
    # sort by the metric provided
    db = db.sort_values(by=metric, ascending=ascending)
    
    # return only key columns
    return db[['id', 'title', metric, 
               'budget_musd', 'revenue_musd', 
               'vote_count', 'vote_average', 'popularity'
              ]].reset_index(drop=True)
```

```python
#using the function to answer the questions earlier 

highest_revenue = rank_movies(df, 'revenue_musd', ascending=False)
highest_budget = rank_movies(df, 'budget_musd', ascending=False)
highest_profit = rank_movies(df, 'profit', ascending=False)
lowest_profit  = rank_movies(df, 'profit', ascending=True)
highest_roi    = rank_movies(df[df['budget_musd'] > 10], 'roi', ascending=False)
lowest_roi     = rank_movies(df[df['budget_musd'] > 10], 'roi', ascending=True)
most_voted     = rank_movies(df, 'vote_count', ascending=False)
highest_rated  = rank_movies(df[df['vote_count'] > 10], 'vote_average', ascending=False)
lowest_rated   = rank_movies(df[df['vote_count'] > 10], 'vote_average', ascending=True)
popular_movie  = rank_movies(df, 'popularity', ascending=False)
```

# Franchise vs. Standalone Movie Performance

```python
#franchise
df['is_franchise'] = df['belongs_to_collection'].apply(lambda x: 0 if x == {} else 1)
franchise_comparison = df.groupby('is_franchise').agg({
    'revenue_musd': 'mean',      
    'roi': 'median',             
    'budget_musd': 'mean',       
    'popularity': 'mean',        
    'vote_average': 'mean'      
}).reset_index()

# Comparison
franchise_comparison['is_franchise'] = franchise_comparison['is_franchise'].map({
    0: "Standalone Movies",
    1: "Franchise Movies"
})

franchise_comparison
```

```python
#most successful movie franchise

franchise_stats = df[df['belongs_to_collection'] != {}].groupby('belongs_to_collection').agg({
    'id': 'count',
    'budget_musd': ['sum', 'mean'],
    'revenue_musd': ['sum', 'mean'],
    'vote_average': 'mean'
}).sort_values(('revenue_musd', 'sum'), ascending=False).head(10)

franchise_stats
```

```python
#most successful director 
director_stats = df.explode('director').groupby('director').agg({
    'id': 'count',
    'revenue_musd': 'sum',
    'vote_average': 'mean'
}).sort_values('revenue_musd', ascending=False).head(10)

director_stats
```
